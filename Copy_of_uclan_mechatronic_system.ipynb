```python
from scipy.io import loadmat
#reading training dataset
url=(r"/content/training_dataset.mat")
data=loadmat(url)
x=data["b"]# b is the name of the matlab file 
print(x)
```

##converting the array to pandas dataframe 

and giving the column names that is 'system_vibration', 'speed_response', 'stability', 'transmission_response', 'power_consumption','shear_stress'and 'class'

```python
training_data=pd.DataFrame(x, columns = ['system_vibration', 'speed_response',
                                         'stability', 'transmission_response', 
                                         'power_consumption','shear_stress','class'])
#np.set_printoptions(formatter={'float': '{:0.2f}'.format})

training_data
```

```python
x=training_data.iloc[:, :-1].values
y=training_data.iloc[:, -1].values
```

#Building a binary classification model using tensorflow

importing the librairies 

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense 
```

```python
model=Sequential()
model.add(Dense(6, input_dim=len(x[0, :]), activation= "relu"))
model.add(Dense(3, activation= "relu"))
#sigmoid activation function is used here because its limit is o and 1 then its perfect for binary  classification
model.add(Dense(1, activation= "sigmoid"))
```

```python
model.summary()
```

##building the model

```python
model.compile(loss="binary_crossentropy", optimizer="rmsprop", metrics=['accuracy'])

#now training the model on the training dataset
model.fit(x,y, epochs=256, verbose=1)
```

##Doing prediction

```python
prediction=model.predict(x)
```

```python
#printing the first 10 prediction 
print(prediction[:10])
```

```python
#printing the truth class 
print(y[:10])
```

##Evaluating the model using the scikit learn library

```python
#importing the standard metrics for evaluating the model

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

###evaluating using accuracy score 

```python
my_accuracy= accuracy_score(y, prediction.round())
print(my_accuracy)
```

The model accuracy is 77%

###evaluating using precision score

```python
precision= precision_score(y,prediction.round())
print(precision)
```

####Evaluating using recall score

```python
recall=recall_score(y, prediction.round())
print(recall)
```

It attained 91% percent accuracy

####Evaluating using f1 score

```python
f1=f1_score(y,prediction.round())
print(f1)
```

it attained 80% percent score
